```python
import pandas as pd

import os
import sys

sys.path.append(os.path.abspath(os.path.join(os.path.dirname(os.getcwd()), "")))


from src.data.load_data import load_raw_data
from src.data.preprocess import preprocess_pipeline
```

```python
df = load_raw_data()
```

```python
df_cleaned = preprocess_pipeline(df)
```

```python
df_cleaned
```

```python
X = df_cleaned.drop(columns=["id", "diagnosis"])
y = df_cleaned["diagnosis"]
```

```python
from src.data.preprocess import (
    compute_f_classif,
    compute_mutual_info,
    compute_logistic_l1_importance,
    compute_rf_importance,
    perform_pca,
)
```

```python
import joblib
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

joblib.dump((X_train, X_test, y_train, y_test), "../data/interim/split_data.pkl")
```

```python
f_scores = compute_f_classif(X_train, y_train)
```

```python
mi_scores = compute_mutual_info(X_train, y_train)
```

```python
l1_scores = compute_logistic_l1_importance(X_train, y_train)
```

```python
rf_scores = compute_rf_importance(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

feature_names = np.array(X_train.columns)


# Sort each score and feature_names in descending order
def sort_scores(scores):
    idx = np.argsort(scores)[::-1]
    return scores[idx], feature_names[idx]


f_scores_sorted, f_names_sorted = sort_scores(f_scores)
mi_scores_sorted, mi_names_sorted = sort_scores(mi_scores)
l1_scores_sorted, l1_names_sorted = sort_scores(l1_scores)
rf_scores_sorted, rf_names_sorted = sort_scores(rf_scores)

fig, ax = plt.subplots(2, 2, figsize=(18, 12))
sns.barplot(x=f_scores_sorted, y=f_names_sorted, ax=ax[0, 0])
ax[0, 0].set_title("ANOVA F-Score")

sns.barplot(x=mi_scores_sorted, y=mi_names_sorted, ax=ax[0, 1])
ax[0, 1].set_title("Mutual Information")

sns.barplot(x=l1_scores_sorted, y=l1_names_sorted, ax=ax[1, 0])
ax[1, 0].set_title("Logistic L1 Importance")

sns.barplot(x=rf_scores_sorted, y=rf_names_sorted, ax=ax[1, 1])
ax[1, 1].set_title("Random Forest Importance")

plt.tight_layout()
plt.show()
```

```python
# Fit PCA on training data
X_train_pca, pca = perform_pca(X_train, n_components=0.95)
explained_var = pca.explained_variance_ratio_

# Transform test data
X_test_pca = pca.transform(X_test)

plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(explained_var), marker="o")
plt.axhline(0.95, color="red", linestyle="--", label="95% Variance")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Explained Variance")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
joblib.dump(pca, "../models/pca.joblib")
joblib.dump((X_train_pca, X_test_pca), "../data/interim/pca_data.pkl")
```

```python
scores_df = pd.DataFrame(
    {
        "F_Score": pd.Series(f_scores),
        "Mutual_Info": pd.Series(mi_scores),
        "L1_Importance": pd.Series(l1_scores),
        "RF_Importance": pd.Series(rf_scores),
    }
)
scores_df.to_csv("../data/interim/feature_scores.csv")
```
